```python
from sklearn.utils import resample

survived = data[data['Survived'] == '0']
died = data[data['Survived'] == '1']

# Upsample class B to match the number of rows in class A
class_B_upsampled = resample(died, replace=True, n_samples = 549, random_state=42)

# Combine the upsampled class B with class A
data = pd.concat([survived, class_B_upsampled])

# Shuffle the DataFrame to mix the rows
data = data.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
dependent = pd.DataFrame(data["Survived"])
```

```python
independent = data.drop(columns = "Survived")
```

```python
model = Sequential ([
     tf.keras.layers.Dense(7, activation = 'relu', name = 'L1'),
     tf.keras.layers.Dense(16, activation = 'relu', name = 'L2'),
     tf.keras.layers.Dense(1, activation = 'sigmoid', name = 'L3')
])
```

```python
independent = np.asarray(independent).astype(np.float32)
dependent = np.asarray(dependent).astype(np.float32)
```

```python
model.compile(
    loss = tf.keras.losses.BinaryCrossentropy(),
    optimizer = tf.keras.optimizers.Adam(learning_rate = 0.01),
    metrics = ['accuracy']
)

model.fit(
    independent, dependent,
    epochs = 60, batch_size = 32
)
```
